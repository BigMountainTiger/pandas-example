```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'id': [1, 2, 3, None, None, None],
    'score': [20, 20, 100, 10, 10, 10]
}).convert_dtypes(dtype_backend='pyarrow')

df
```

<h1>np.full(len(df), True)</h1>
<h4>1. It can be used to create an always true series</h4>
<h4>2. It can then be combined with other condictions</h4>

```python
always_true = np.full(len(df), True)

df.loc[always_true, :]
```

<h4>Combined with other condictions</h4>

```python
always_true = np.full(len(df), True)

df.loc[always_true & (df['score'] > 20), :]
```
