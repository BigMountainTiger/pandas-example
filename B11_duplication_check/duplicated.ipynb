```python
import pandas as pd

df = pd.DataFrame({
    'id': ['English', 'English', 'Math', 'Math', 'Science'],
    'student': ['Song', 'Song', 'Trump', 'Trump', 'Biden']
})

df
```

<h1>duplicated() output a Series where True indicates a duplicate row (after the first occurrence)</h1>

<h4>Controlling which duplicates are marked</h4>
<h4>
    <ol>
        <li>keep='first' (default): Marks duplicates as True except for the first occurrence</li>
        <li>keep='last': Marks duplicates as True except for the last occurrence</li>
        <li>keep=False: Marks all occurrences of duplicates as True</li>
    </ol>
</h4>

```python
df.duplicated(keep=False)
```

<h1>subset=[] specifies the columns to check for the duplications</h1>

```python
df.duplicated(keep=False,subset=['id', 'student'])
```

<h1>Get the duplicate rows</h1>

```python
df.loc[df.duplicated(keep=False,subset=['id', 'student']), ['id', 'student']]
```

<h1>Get unique from the duplicated</h1>

<h4>
    The drop_duplicates() method has similar parameters as duplicated()
</h4>
<h4>
    <ol>
        <li>keep - default keep the first</li>
        <li>subset - default to check all the columns</li>
    </ol>
</h4>

```python
df.loc[df.duplicated(keep=False,subset=['id', 'student']), ['id', 'student']].drop_duplicates()
```

```python
len(df.loc[df.duplicated(keep=False,subset=['id', 'student']), ['id', 'student']].copy().drop_duplicates())
```
