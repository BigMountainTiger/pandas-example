```python
import pandas as pd
```

```python
df = pd.read_csv('https://covid19.who.int/WHO-COVID-19-global-data.csv')

df['Country_code'] = df['Country_code'].astype('category')
df['Country'] = df['Country'].astype('category')
df['WHO_region'] = df['WHO_region'].astype('category')
df['Date_reported'] = pd.to_datetime(df['Date_reported']).astype('category')

df.head(1)
```

```python
# agg without group
df.agg({'New_cases': ['min', 'max', 'mean']})
```

```python
new_cases = df['New_cases']
result = {
    'min': new_cases.min(),
    'max': new_cases.max(),
    'mean': new_cases.mean()
}

result
```

```python
ag = df.groupby(['Country_code']).agg({'New_cases': ['min', 'max', 'mean']})
ag.loc[['US']]
```

```python
# Access the details
ag['New_cases'][['min', 'max', 'mean']].loc['US']
```
