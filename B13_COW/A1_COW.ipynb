<h1>Avoid chained assignment</h1>

```python
df = pd.DataFrame({
    'id': [1, 2],
    'student': ['Song', 'Trump']
})

df.loc[df['id'] == 1]['score'] = 100
```

<h1>Syntax sugers in pandas cause confusions</h1>
<h4>The following works as the assignment is on a new data frame</h4>

```python
df = pd.DataFrame({
    'id': [1, 2],
    'student': ['Song', 'Trump']
})

df1 = df.loc[df['id'] == 1]
df1['score'] = 100

df1
```
