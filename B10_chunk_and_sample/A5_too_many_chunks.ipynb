```python
import pandas as pd
import numpy as np
import random

D = {
    'class': ['English', 'English', 'Math', 'Math'],
    'student': ['Song', 'Trump', 'Song', 'Trump'],
    'score': [100, 10, 20, 90]
}
df = pd.DataFrame(D)
df
```

<h1>If the number of the chunks more than the number of rows in the original dataframe, empty chunks can be generated</h1>
<h4>If necessary the chunks (list of the sub-dateframes) can get random.shuffle() to create randomness</h4>

```python
no_of_chunks = 6

dfs = []
indices = np.array_split(df.index, no_of_chunks)
for chunk_idx in indices:
    dfs.append(df.loc[chunk_idx].copy())
    random.shuffle(dfs)

dfs
```

<h1>When concat together, it is OK the list has empty dateframes</h1>

```python
pd.concat(dfs, ignore_index=True)
```
