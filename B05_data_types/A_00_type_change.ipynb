```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'class': ['English', 'Math'],
    'score': [None, 80]
})
df1['score'] = df1['score'].astype('Int32')

df1.info()
df1
```

```python
df2 = pd.DataFrame({
    'class': ['Science', 'Sports'],
    'score': [None, None]
})
df2['score'] = df2['score'].astype('Int32')

df2.info()
df2
```

<h1>It is better to have the same type for the column to concat()</h1>
<h1>If the same column from different frames have different types, the more generic one will be the type of the result frame</h1>

```python
df = pd.concat([df1, df2])
df.info()
df
```
