```python
import pandas as pd
```

<h1>Fill None with Default value, True for example</h1>

```python
df = pd.DataFrame({'bool_col': [True, False, True, None, None]})
df.convert_dtypes()

df.info()
df
```

<h1>If None exists in the values, the cast will fail</h1>

```python
df = pd.DataFrame({'bool_col': [True, False, True, None, None]})
df.convert_dtypes()

df['bool_col'] = df['bool_col'].astype('string').str.strip().str.lower().map({'true': 1, 'false': 0}).astype('Int64')
df
```

```python
df.info()
```

<h1>Use map() seems to be a more fault tolerant method</h1>

```python
df = pd.DataFrame({'bool_col': [True, False, True, None, None, 'Non-mappable']})
df
```

```python
df = pd.DataFrame({'bool_col': [True, False, True, None, None, 'Non-mappable']})
df['bool_col'] = df['bool_col'].map({True: 1, False: 0})
df['bool_col'] = df['bool_col'].astype('Int64')
df.convert_dtypes()
df
```

```python
df.info()
```
