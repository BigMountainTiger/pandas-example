```python
import pandas as pd
```

<h1>pd.isnull() and pd.isna() are aliases</h1>

<h4>None, pd.NA, pd.NaT all recognized as NULL</h4>

```python
pd.isnull(None)
```

```python
pd.isnull(pd.NA)
```

```python
pd.isnull(pd.NaT)
```

```python
pd.isnull('a string')
```

```python
pd.isna(None)
```

```python
pd.isna(pd.NA)
```

```python
pd.isna(pd.NaT)
```

```python
pd.isna('a string')
```
