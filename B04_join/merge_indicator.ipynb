<h1>With indicator=True a column named _merge will be created</h1>

```python
import pandas as pd

student = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Song', 'Biden', 'Trump'],
})

student_score = pd.DataFrame({
    'student_id': [1, 3, 100],
    'score': [100, 20, 90],
})


student.merge(student_score, how='outer', left_on=['id'], right_on=['student_id'], indicator=True)
```

<h1>If the indicator column name exist an run-time exception</h1>

```python
import pandas as pd

student = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Song', 'Biden', 'Trump'],
    '_merge': ['ok', 'ok', 'ok']
})

student_score = pd.DataFrame({
    'student_id': [1, 3, 100],
    'score': [100, 20, 90],
})


student.merge(student_score, how='outer', left_on=['id'], right_on=['student_id'], indicator=True)
```

<h1>Need to specify a different column name as the indicator</h1>

```python
import pandas as pd

student = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Song', 'Biden', 'Trump'],
    '_merge': ['ok', 'ok', 'ok']
})

student_score = pd.DataFrame({
    'student_id': [1, 3, 100],
    'score': [100, 20, 90],
})


student.merge(student_score, how='outer', left_on=['id'], right_on=['student_id'], indicator='_merge_indicator')
```
