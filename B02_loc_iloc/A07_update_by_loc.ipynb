```python
import pandas as pd
```

<h1>1. Student dateframe</h1>

```python
df_student = pd.DataFrame({
    'id': [0, 1, 2, 3],
    'name': ['Song', 'trump', 'Biden', 'Potato'],
    'score': [None, None, None, None]
})

df_student
```

<h1>2. Score dataframe</h1>

```python
df_score = pd.DataFrame({
    'name': ['Song', 'Potato'],
    'score': [50, 90]
})

df_score
```

<h1>3. merge two dataframes with indicator=True</h1>

```python
df_merged = df_student.merge(df_score, on='name', how='left', suffixes=('', '_r'), indicator=True)
df_merged = df_merged.convert_dtypes()

df_merged
```

<h1>4. Update the merged rows if df_merged['_merge'] == 'both'</h1>

```python
df_merged.loc[df_merged['_merge'] == 'both', 'score'] = df_merged['score_r']
df_merged
```

```python
df_merged.loc[:, list(df_student.columns)]
```
